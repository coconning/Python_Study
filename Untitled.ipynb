```python
import pandas as pd

data=pd.read_excel("./테스트1.xlsx", engine="openpyxl").reset_index(drop=True)
data.index=data.index+2
print(data.shape)
data.head()
```

```python
start_idx = data.index[0]
end_idx   = data.index[-1]
cols    = data.columns
cut_num = 100

for i in range(start_idx,end_idx, cut_num):
    print(i-2,i+cut_num-2)
    data.loc[i:i+cut_num-1,cols].to_excel(f"./test_folder/test_{i-2}_{i+cut_num-2}.xlsx", index=False)
```
