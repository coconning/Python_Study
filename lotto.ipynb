```python
from sklearn.dummy import DummyClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import numpy as np
import pandas as pd

dataset=make_classification(n_samples=100000, n_features=3, n_informative=3, n_redundant=0, n_classes=2, weights=(0.18,0.82), flip_y=0)
dataset=pd.DataFrame(data={"X1":dataset[0][:,0],"X2":dataset[0][:,1],"X3":dataset[0][:,2], "y":dataset[1]})

X=dataset[["X1","X2","X3"]]
y=dataset.y

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

clfs=[(f"Dum{i}", DummyClassifier(strategy="stratified", )) for i in range(9)]

for n, clf in clfs:
    clf.fit(X_train, y_train)
    print(accuracy_score(clf.predict(X_test), y_test))    

vc=VotingClassifier(estimators=clfs, )

vc.fit(X_train, y_train)
print("Voting", accuracy_score(vc.predict(X_test), y_test))
```
